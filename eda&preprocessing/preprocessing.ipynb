```python
#importing dependencies
import os 
import pandas as pd
import numpy as np
import nltk
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import re
from nltk.corpus import stopwords

```

```python
nltk.download('stopwords')
nltk.download('punktab')
```

```python
data = pd.read_csv(r'C:\Users\IKE\Desktop\Human_VS_AI\eda&preprocessing\AI VS Human.csv')
data.head()
```

```python
#preprocess the text
def preprocessor(text):
    #convert text to lower case
    text = text.lower()

    # remove special characters
    text = text = re.sub(r'[^a-z\s]', '', text)

    # tokenize text
    token = word_tokenize(text)

    # stopwords
    stop_words = set(stopwords.words('english'))

    tokens = [word for word in token if word not in stop_words and len(word) >1]

    # apply lemmitization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens]

    return ' '.join(tokens)

data['text_preprocessed'] = data['text_content'].apply(preprocessor)
data.head(4)
```

```python
preprocessed_data = data.copy()

# saving the preprocessed data
preprocessed_data.to_csv('Cleaned_data.csv',index=False)
```
